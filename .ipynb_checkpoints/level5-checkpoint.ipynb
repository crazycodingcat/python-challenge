```python
import pickle
```

```python
data_path = "/home/xuan/Desktop/Workspace/PyChallenge/data/level5.p"
data = pickle.load(open(data_path, 'rb'))
```

```python
data
```

```python
#good solution
for line in data:
    print (''.join(c*n for c, n in line)) #generator
```

```python
#another good solution
for line in data:
    print "".join(map(lambda pair: pair[0]*pair[1], line))
```

```python
#good solution
for line in data:
    line_item = [c*n for c, n in line]
    print ''.join(line_item)
```

```python
#bad solution
for line in data:
    line_string = ''
    for item in line:
        line_string += item[0]*item[1] 
    print line_string
```
